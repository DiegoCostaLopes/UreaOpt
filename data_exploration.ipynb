```python
# let's check if there are duplicate names within the same state
# there are not. 

ref_col = 'region_name'
unique_states = df[ref_col].unique()

for state in unique_states:
    data = df.loc[df[ref_col] == state]
    dupes = data['name'].loc[data['name'].duplicated()]
    if dupes.empty is False:
        display(dupes)  
    
```
