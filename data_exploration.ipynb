```python
# let's check if there are duplicate names within the same state
# there are not. 

ref_col = 'region_name'
unique_states = df[ref_col].unique()

for state in unique_states:
    data = df.loc[df[ref_col] == state]
    dupes = data['name'].loc[data['name'].duplicated()]
    if dupes.empty is False:
        display(dupes)
    
```

```python
data = pd.read_excel('data/processed/soy.xlsx')
data = data.loc[data['Tipo região'] == 'Município']
data
```

```python
def get_id(series):
    # gets id based on name
    id = df.loc[df['name'] == series['Local'], 'id_city']
    # id = df.at[df['name'] == series['Local'], 'id_city']
    # display(id)
    if id.empty:
        return np.nan
    else:
        return id.values[0]

# bla = get_id(data.loc[4])
# bla

data['id'] = data.apply(get_id, axis=1)
data
```
